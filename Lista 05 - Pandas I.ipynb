## QUESTÃO 5

Crie uma nova coluna no Dataframe "df" com o nome "Nota_Geral" que represente a média da nota de cada aluno nas 3 disciplinas. Utilize esta nova coluna para ordenar os alunos de acordo com sua nota de forma que o aluno com maior nota geral esteja na primeira linha do DataFrame.

```python
df["Nota_Geral"] = (df['Nota_Leitura'] + df['Nota_Matemática'] + df["Nota_Redação"])/3

df = df.sort_values(by = "Nota_Geral", ascending=False, ignore_index= True)
display(df)
```

## QUESTÃO 6

a) Conte quantos dos alunos pertencem a cada gênero.

```python
#SEU CÓDIGO COMEÇA AQUI

mulheres = 0
homens = 0

for valores in df["Gênero"]:
    if valores == "female":
        mulheres = mulheres + 1
    else:
            homens = homens + 1

print(f' A quantidade de mulheres é {mulheres} e a de homens {homens}')
```

b) Conte quantos dos alunos pertencem a cada Raça/Etnia.

```python
#SEU CÓDIGO COMEÇA AQUI
group_A = 0
group_B = 0
group_C = 0
group_D = 0
group_E = 0

for valores in df["Raça/Etnia"]:
    if valores == "group A":
        group_A = group_A + 1
    elif valores == "group B":
        group_B = group_B + 1
    elif valores == "group C":
        group_C = group_C + 1
    elif valores == "group D":
        group_D = group_D + 1
    elif valores == "group E":
        group_E = group_E + 1


print(f'A quantidade de membros dos grupos A, B, C, D e E é respectivamente: {group_A}, {group_B}, {group_C}, {group_D}, {group_E}')
```

c) Conte quantos dos alunos realizaram ou não algum curso preparatório.

```python
#SEU CÓDIGO COMEÇA AQUI
fez = 0
não_fez = 0

for valores in df["Curso_Prep"]:
    if valores == "completed":
        fez = fez + 1
    else:
        if valores == "none":
            não_fez = não_fez + 1

print(f'A quantidade de pessoas que realizaram um curso preparatório foram de {fez}, já as que não fizeram foi de {não_fez}')


```

## QUESTÃO 8

Neste exercício deseja-se observar qual das disciplinas examindas (Matemática, Leitura, Redação) é mais influênciada por cada um dos fatores qualitivos (Gênero, Raça/Etnia, Realização de Curso Preparatório). Para isso, agrupe a base de dados para cada um dos fatores qualitativos e calcule a diferença entre a maior e menor nota média em cada uma das disciplinas. A disciplina que possuir maior diferença entre as notas médias será considerada a mais influênciada.

a) Qual disciplina possui a maior diferença entre as notas médias de homens e mulheres.

```python
#SEU CÓDIGO COMEÇA AQUI
agrupamento_por_gênero = df.groupby("Gênero")[["Nota_Matemática", "Nota_Leitura","Nota_Redação"]].mean()
agrupamento_por_gênero

display(agrupamento_por_gênero)

dif_mat = agrupamento_por_gênero.iloc[1,0] - agrupamento_por_gênero.iloc[0,0]
dif_leit = agrupamento_por_gênero.iloc[0,1] - agrupamento_por_gênero.iloc[1,1]
dif_red = agrupamento_por_gênero.iloc[0,2] - agrupamento_por_gênero.iloc[1,2]

print(dif_mat)
print(dif_leit)
print(dif_red)

print("A maior diferença nas notas foi em Nota_Redação")
```

b) Qual disciplina possui a maior diferença entre a maior e menor nota média dentre as diferentes Raças/Etnias.

```python
#SEU CÓDIGO COMEÇA AQUI
agrupamento_por_raça_etnia = df.groupby("Raça/Etnia")[["Nota_Matemática", "Nota_Leitura","Nota_Redação"]].mean()
display(agrupamento_por_raça_etnia)

dif_mat_rac = agrupamento_por_raça_etnia.iloc[4,0] - agrupamento_por_raça_etnia.iloc[0,0]
dif_leit_rac = agrupamento_por_raça_etnia.iloc[4,1] - agrupamento_por_raça_etnia.iloc[0,1]
dif_red_rac = agrupamento_por_raça_etnia.iloc[4,2] - agrupamento_por_raça_etnia.iloc[0,2]

print(dif_mat_rac)
print(dif_leit_rac)
print(dif_red_rac)

print(f'A nota com maior diferença entre as diferentes raças é em Nota_Matemática')
```

c) Qual disciplina possuir a maior diferença entre as notas médias daqueles que realizaram ou não cursos preparatórios.

```python
#SEU CÓDIGO COMEÇA AQUI
agrupamento_por_curso_prep = df.groupby("Curso_Prep")[["Nota_Matemática", "Nota_Leitura","Nota_Redação"]].mean()

display(agrupamento_por_curso_prep)

dif_curso_prep_mat = agrupamento_por_curso_prep.iloc[0,0] - agrupamento_por_curso_prep.iloc[1,0]
dif_curso_prep_leit = agrupamento_por_curso_prep.iloc[0,1] - agrupamento_por_curso_prep.iloc[1,1]
dif_curso_prep_red = agrupamento_por_curso_prep.iloc[0,2] - agrupamento_por_curso_prep.iloc[1,2]

print(dif_curso_prep_mat)
print(dif_curso_prep_leit)
print(dif_curso_prep_red)

print("A maior diferença é em Nota_Redação")
```
